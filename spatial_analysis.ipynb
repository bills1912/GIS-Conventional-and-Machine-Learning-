```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
geom_ds = pd.read_csv('dataset-limpo.csv')
```

```python
geom_ds.head()
```

```python
print("Ukuran data:", geom_ds.shape)
geom_ds.isnull().sum()
```

```python
geom_ds_clean = geom_ds.dropna(axis=1)
```

```python
print("Ukuran data:", geom_ds_clean.shape)
geom_ds_clean.isnull().sum()
```

```python
geometry = [Point(xy) for xy in zip(geom_ds_clean['longitude'], geom_ds_clean['latitude'])]
```

```python
geo_df = gpd.GeoDataFrame(geom_ds_clean, crs=dict(init=['epsg:4326']), geometry=geometry)
```

```python
geo_mask = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
```

```python
geo_mask.plot()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
geo_mask.plot(ax=ax, alpha=0.4, color='grey')
geo_df.plot(ax=ax, markersize=15, color='green', marker='o')
plt.show()
```
